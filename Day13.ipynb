**4.Checking For Null Values**

```python
print('checking for null values......')
is_null = pd.isnull(dataset)
print(is_null)
print('calculating total null values in dataset')
null_columns=dataset.columns[dataset.isnull().any()]
dataset[null_columns].isnull().sum()
print("printing null value columns and rows......")
print(dataset[dataset.isnull().any(axis = 1)][null_columns].head())
```

**5.Checking For Duplicate Values**

```python
print('checking for duplicated values.....')
duplicated_values = pd.DataFrame(dataset)
print(duplicated_values.duplicated(subset=None, keep='first'))
```
